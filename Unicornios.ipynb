```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
base_dados = pd.read_csv('/unicorns till sep 2022.csv')

base_dados.shape
```

```python
base_dados.head()
```

```python
base_dados.columns
```

```python
base_dados.rename(columns={
                           'Company' :'Empresa',
                           'Valuation ($B)' :'valor($)' ,
                           'data de adesÃ£o' :'Data de adesÃ£o',
                           'Country' :'Pais',
                           'City' : 'Cidade',
                           'Industry' : 'Setor' ,
                           'Investors' : 'Investidores'  }, inplace=True )
```

```python
base_dados.info()
```

```python
base_dados.isnull().sum()
```

```python
plt.figure(figsize = (15,6))
plt.title('Analisando Campos Nulos')
sns.heatmap(base_dados.isnull(), cbar=False);
```

```python
base_dados.nunique()
```

```python
base_dados['Setor'].unique()
```

```python
base_dados['Setor'].value_counts()
```

```python
base_dados['Setor'].value_counts(normalize= True)
```

```python
plt.figure(figsize = (15,5))
plt.title('Analisando os Setores')
plt.xticks(rotation = 45, ha= 'right')
plt.bar(base_dados['Setor'].value_counts().index, base_dados['Setor'].value_counts())
```

```python
analisando = round ( base_dados['Pais'].value_counts(normalize= True) * 100,1)
```

```python

plt.figure(figsize = (15,6))
plt.title('AnalisandoPaises')
plt.pie(
        analisando,
        labels = analisando.index,
        shadow = True,
        startangle = 90,
        autopct= '%1.1f%%'
        );
```

```python

plt.figure(figsize = (15,6))
plt.title('Analisando Paises - TOP 10')
plt.pie(
        analisando.head(10),
        labels = analisando.index [0:10],
        shadow = True,
        startangle = 90,
        autopct= '%1.1f%%'
        );
```
